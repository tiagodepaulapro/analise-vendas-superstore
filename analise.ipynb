```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Análise de Vendas - Superstore

## Objetivo da Análise

Este projeto tem como objetivo explorar dados de vendas de uma empresa do setor varejista, com foco na compreensão de padrões de faturamento e lucratividade.

A proposta é aplicar conceitos iniciais de análise de dados, buscando identificar quais categorias apresentam melhor desempenho e onde podem existir oportunidades de melhoria.

Este projeto foi desenvolvido como parte do meu processo de aprendizado na área de dados.

---

## Ferramentas Utilizadas

* Python
* Pandas
* Matplotlib



```python
df = pd.read_csv("Sample - Superstore.csv", encoding="latin1")

df.head()
```

```python
vendas_categoria = df.groupby("Category")["Sales"].sum()

vendas_categoria
```

```python
vendas_categoria = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(8,5))

vendas_categoria.plot(kind="bar")

plt.title("Total de Vendas por Categoria")
plt.ylabel("Vendas")
plt.xlabel("Categoria")

plt.tight_layout()

plt.savefig("grafico_vendas_categoria.png")
plt.show()
```

## Análise de Vendas por Categoria

A análise mostra que a categoria Technology apresenta o maior volume de vendas, seguida por Furniture e Office Supplies.

Isso indica que os produtos de tecnologia têm maior participação no faturamento, podendo ser considerados estratégicos para o negócio.

Neste ponto, a análise foca apenas em volume de vendas, sem considerar ainda a lucratividade.

```python
df.groupby("Category")["Profit"].sum()
```

```python
lucro_categoria = df.groupby("Category")["Profit"].sum()

lucro_categoria.plot(kind="bar")
plt.title("Lucro por Categoria")
plt.ylabel("Lucro")
plt.show()
```

## Análise de Lucro por Categoria

Ao analisar o lucro, observa-se que a categoria Technology continua com o melhor desempenho.

Por outro lado, a categoria Furniture apresenta um lucro significativamente menor, mesmo tendo um bom volume de vendas.

Isso sugere que nem sempre vender mais significa lucrar mais, sendo importante analisar também os custos envolvidos.

```python
lucro_subcategoria = df[df["Category"] == "Furniture"] \
    .groupby("Sub-Category")["Profit"].sum() \
    .sort_values()

lucro_subcategoria
```

```python
lucro_subcategoria = df[df["Category"] == "Furniture"] \
    .groupby("Sub-Category")["Profit"].sum()

lucro_subcategoria.plot(kind="bar")
plt.title("Lucro por Subcategoria - Furniture")
plt.ylabel("Lucro")
plt.show()
```

## Análise de Subcategorias - Furniture

Para entender melhor o desempenho da categoria Furniture, foi feita uma análise mais detalhada por subcategoria.

Os dados mostram que produtos como Tables e Bookcases apresentam prejuízo, enquanto Chairs e Furnishings têm resultados positivos.

Isso indica que o problema da categoria pode estar concentrado em itens específicos, e não no grupo como um todo.

## Conclusão Geral da Análise

A análise permitiu identificar que a categoria Technology é a principal responsável pelo faturamento e lucro da empresa.

Já a categoria Furniture, apesar de apresentar boas vendas, possui baixa eficiência em termos de lucratividade, impactada por subcategorias com prejuízo.

Esse tipo de análise mostra a importância de avaliar não apenas o volume de vendas, mas também a qualidade do resultado financeiro.

Este projeto representa uma aplicação prática dos conceitos iniciais de análise de dados, com foco na interpretação e geração de insights a partir dos dados.
